```python
df_cr.new = df_cr.rename(index=str,columns={'movie_id':'id'})
df_movies = df_mov.merge(df_cr.new,on='id')
```

```python
df_movies.head()
```

```python
df = df_movies.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
df
```

```python
df.info()
```

```python
#average rating for the movie
R = df['vote_average']

#no of vote for the movie
V = df['vote_count']

#mean vote across the whole report
C = df['vote_average'].mean()

#minimum vote required
M = df['vote_count'].quantile(0.70)
```

```python
df['weighted_avg'] = ((R*V)+(C*M))/(V+M) 
```

```python
df.head()
```

```python
#sort dataframe on the basis of weighted_avg

df_sorted = df.sort_values('weighted_avg',ascending=False)
df_sorted[['original_title','popularity','vote_average','weighted_avg']].head()
```

```python
#sort dataframe on the basis of popularity

df_popularity = df.sort_values('popularity',ascending=False)
df_popularity[['original_title','popularity','vote_average','weighted_avg']].head()
```

```python
#recommendation on the basis of both popularity and weighted avg

from sklearn.preprocessing import MinMaxScaler

scalaing = MinMaxScaler()
scaled_df= scalaing.fit_transform(df[['popularity','weighted_avg']])

df_normalized = pd.DataFrame(scaled_df,columns=['popularity','weighted_avg'])
```

```python
df_normalized.head()
```

```python
df[['popularity_normalized','weighted_avg_normalized']] = df_normalized
```

```python
df.head(1)
```

```python
df['score'] = df['popularity_normalized']*0.5 + df['weighted_avg_normalized']*0.5
df_score    = df.sort_values(['score'],ascending=False)
df_score[['original_title','popularity_normalized','weighted_avg_normalized','score']]
```
