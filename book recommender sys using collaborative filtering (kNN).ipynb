```python
#setting threshold for the rating counts 

counts1 = ratings['User-ID'].value_counts()
ratings = ratings[ratings['User-ID'].isin(counts1[counts1 >= 200].index)]
counts = ratings['Book-Rating'].value_counts()
ratings = ratings[ratings['Book-Rating'].isin(counts[counts >= 100].index)]
```

```python
#collaborative filtering using kNN

combine_book_rating = pd.merge(ratings,books,on='ISBN')
cols = ['Year-Of-Publication', 'Publisher', 'Book-Author', 'Image-URL-M', 'Image-URL-L', 'Image-URL-S']

combine_book_rating = combine_book_rating.drop(cols,axis=1)

combine_book_rating.head()
```

```python
combine_book_rating.info()
```

```python
combine_book_rating = combine_book_rating.dropna(axis=0,subset=['Book-Title'])

combine_book_rating.info()
```

```python
book_rating_count = (combine_book_rating.groupby('Book-Title')['Book-Rating'].count().reset_index().rename(columns={'Book-Rating':'BookRatingCount'}))
```

```python
book_rating_count.sort_values('BookRatingCount',ascending=False)
```

```python
combine_book_rating.head()
```

```python
rating_with_rating_count = book_rating_count.merge(combine_book_rating,left_on='Book-Title',right_on='Book-Title',how='left')
```

```python
rating_with_rating_count.sort_values('BookRatingCount',ascending=False)
```

```python
popularity_threshold =50

popular_rated = rating_with_rating_count.query('BookRatingCount >= @popularity_threshold')
```

```python
popular_rated.head()
```

```python
popular_rated.shape
```

```python
from scipy.sparse import csr_matrix

popular_rated = popular_rated.drop_duplicates(['Book-Title','User-ID'])
book_user_rating_pivot = popular_rated.pivot(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating').fillna(0)

book_user_rating_matrix = csr_matrix(book_user_rating_pivot.values)
```

```python
book_user_rating_pivot
```

```python
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(book_user_rating_matrix)
```

```python
#randomly selecting a index no# of a book
query_index = np.random.choice(book_user_rating_pivot.shape[0])
print(query_index)
```

```python
#get distance and indices params using knn 
distances, indices = model_knn.kneighbors(book_user_rating_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
book_user_rating_pivot.index[query_index]
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(book_user_rating_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, book_user_rating_pivot.index[indices.flatten()[i]], distances.flatten()[i]))

```
